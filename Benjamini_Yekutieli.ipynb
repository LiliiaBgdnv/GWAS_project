```python
import statsmodels.stats.multitest as stat
import pandas as pd
import matplotlib.pyplot as plt
from qmplot import manhattanplot
```

## BLINK complex trait

```python
df_soy_complex_b = pd.read_table('GAPIT/BLINK_complex/GAPIT.Association.GWAS_Results.BLINK.pheno.csv', sep=',')
df_soy_complex_b['P.value_BY'] = stat.fdrcorrection(pvals=df_soy_complex_b['P.value'], 
                                                                           alpha=0.05, 
                                                                           method='negcorr', 
                                                                           is_sorted=False)[1]
if __name__ == "__main__":
    # generate manhattan plot and set an output file.
    fig, ax = plt.subplots(figsize=(10, 7), facecolor="w", edgecolor="k")
    ax = manhattanplot(data=df_soy_complex_b,
                       suggestiveline=0.05,
                       marker=".",
                       chrom="Chr",
                       pos="Pos",
                       title="BLINK complex trait",
                       sign_marker_color="r",
                       snp="SNP",
                       xlabel="Chr",
                       pv='P.value_BY',
                       is_annotate_topsnp=True,
                       text_kws={"fontsize": 200},
                       ax=ax)
```

Let's see which SNPs have a P-value with Benjamini-Yetkutieli correction less than 0.05

```python
df_soy_complex_b.loc[df_soy_complex_b['P.value_BY'] < 0.05]
```
